```python
def filter_2020(x):
    if x[0] == '=':
        return int(x.split('=')[1])
    else:
        return int(x)
```

```python
data['rank_2020']=data['rank_2020'].apply(lambda x :filter_2020(x))
```

```python
data['rank_2019'].values
```

```python
def filter_2019(x):
    if x[0][0] in '0123456789':
        try:
            x[0] = int(x[0])
            return x[0]
        except:
            return int(np.mean([int(i) for i in (x[0].split('–'))]))
    elif x[0] == 'NR':
        return x[1]
    else:
        return int(x[0].split('=')[1])
            
```

```python
data[['rank_2019','rank_2020']].apply(filter_2019,axis = 1).values
```

```python
filter_2019(data['rank_2019'].iloc[165])
```

```python
filter_2019(data['rank_2019'].iloc[1])
```

```python
data['rank_2019'] = data[['rank_2019','rank_2020']].apply(filter_2019,axis = 1)
```

```python
data.groupby('Country').mean().sort_values('rank_2020',ascending=True)
```

```python
def diff(x):
    return x[0]-x[1]
```

```python
data.drop('diff',inplace=True,axis=1)
```

```python
data['Rank_Improved']= data[['rank_2019','rank_2020']].apply(diff,axis= 1)
```

```python
data.sort_values('Rank_Improved',ascending=False) 
```

```python
sorted_data_frame = data.sort_values(['Rank_Improved'], ascending=False)
```

```python
grouped_data_frame = sorted_data_frame.groupby('Country')
```

```python
rankData = pd.concat([data, ranks], axis = 1).sort_values(['Country','rank_for_most_improvement'])
```
